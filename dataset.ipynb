```python
import os
import csv
import pandas as pd
```

```python
for root, dirs, files in os.walk('/home/aditya/Repo/HurricaneTrajectoryRNN/data') :
    pass
rows = []
```

```python
with open('/home/aditya/Repo/HurricaneTrajectoryRNN/main.csv', 'w') as main_csv_file :
    data_writer = csv.writer(main_csv_file)
    for file in files:
        with open(os.path.join('/home/aditya/Repo/HurricaneTrajectoryRNN/data', file)) as csv_file :
            data_reader = csv.reader(csv_file)
            head = next(data_reader)
            for row in data_reader :
                row.append(file[:-4])
                rows.append(row)
    data_writer.writerows(rows)
```

```python
df = pd.read_csv('main.csv')
```

```python
df.set_axis(['LATITUDE', 'LONGITUDE', 'TIME', 'WIND SPEED', 'PRESSURE', 'STATUS', 'NAME'], axis = 1, inplace = True)
```
